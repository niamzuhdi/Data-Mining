```python
x=no3.iloc[0:550068,4:12]
x1=x.drop('City_Category',1)
x2=x1.drop('Stay_In_Current_City_Years',1)
```

```python
#normalization
from sklearn.preprocessing import normalize
xnorm=normalize(x2,norm='l2') #default norm L2, other are L1 dan Max
xnorm=pd.DataFrame(xnorm)
xnorm.tail()
```

```python
#3sigmoid
xsigmoidal=(1-2.718281828**(-xnorm))/(1+2.718281828**(-xnorm))
xsigmoidal=pd.DataFrame(xsigmoidal)
xsigmoidal.tail()
```

```python
ets3=xsigmoidal.describe()
ets3.to_csv('no3.csv', sep=',')
```
