```python
import numpy as np
import pandas as pd
from apyori import apriori 
```

```python
ets = pd.read_csv('online_retail.csv')
ets.head()
```

```python
np.sum(ets.isnull()) #menemukan data yang kosong
```

# Membersikan Data karena Description yang kosong dan tidak ada penjualan serta juga ada yang terdapat kredit (simbol c pada transaksi)

```python
ets['UnitPrice'] = ets['UnitPrice'].astype('str')
ets = ets[~ets['UnitPrice'].str.contains('0')]
```

```python
np.sum(ets.isnull())
```

```python
#mengambil data pada InvoiceNo untuk mengelompokkan barang-barang menjadi satu per transaksi
x = ets.groupby('InvoiceNo')
transaksi = []
for i,j in x:
    transaksi.append(list(j['Description'].map(str)))
```

```python
pd.DataFrame(transaksi)
```

```python
data=pd.DataFrame(transaksi)
```

```python
data.shape
```

```python
records=[]
for i in range (0, 22465):
    records.append([str(data.values[i,j]) for j in range (0,923)])
```

```python
association_rules=apriori(records, min_support=0.0045,
                         min_confidence=0.2,
                         min_lift=3, min_length=2)
association_result=list(association_rules)
```

```python
association_result[0]
```

```python
for item in association_result:
#first index of the inner list
#contains base item and add item
    pair = item[0]
    items = [x for x in pair]
    print("rule: " + items [0] + " ->" + items [1])
    
    #second index of the inner list
    print("support: " + str(item[1]))
    
    #third index of the list located at 10th
    #of the third index of the inner list
    
    print("confidence: " + str(item[2][0][2]))
    print("lift: " + str(item[2][0][3]))
    print("=====================================")
```
