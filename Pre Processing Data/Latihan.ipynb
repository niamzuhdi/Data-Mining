```python
import pandas as pd
import numpy as np
```

```python
movie=pd.read_csv("movie_metadata.csv")
```

```python
movie.head()
```

```python
movie.describe() #Descriptive statistics
```

```python
np.sum(movie.isnull()) #menemukan data yang kosong
```

```python
movie['budget']=movie['budget'].fillna(3.975262e+07)
np.sum(movie.isnull())
```

```python
movie['gross']=movie['gross'].fillna(4.846841e+07)
np.sum(movie.isnull())
```

```python
movie['director_facebook_likes']=movie['director_facebook_likes'].fillna(686.621709)
np.sum(movie.isnull())
```

```python
movie['num_critic_for_reviews']=movie['num_critic_for_reviews'].fillna((movie['num_critic_for_reviews'].mean()))
movie['actor_3_facebook_likes']=movie['actor_3_facebook_likes'].fillna((movie['actor_3_facebook_likes'].mean()))
movie['actor_2_facebook_likes']=movie['actor_2_facebook_likes'].fillna((movie['actor_2_facebook_likes'].mean()))
movie['actor_1_facebook_likes']=movie['actor_1_facebook_likes'].fillna((movie['actor_1_facebook_likes'].mean()))
movie['num_user_for_reviews']=movie['num_user_for_reviews'].fillna((movie['num_user_for_reviews'].mean()))
movie['facenumber_in_poster']=movie['facenumber_in_poster'].fillna((movie['facenumber_in_poster'].mean()))
movie['duration']=movie['duration'].fillna((movie['duration'].mean()))
np.sum(movie.isnull())
```

```python
from scipy.stats import mode

#Fill missing value with mode for qualitative/categorical data

movie['country']=movie['country'].fillna(movie['country']).dropna().mode()[0]
np.sum(movie.isnull())
```

```python
#mengatasi outlier
from scipy import stats
movie1=movie.drop('country',1)
z= np.abs(stats.zscore(movie1._get_numeric_data()))
print(z)
movie=movie[(z<3).all(axis=1)]
print(movie.shape)
```
